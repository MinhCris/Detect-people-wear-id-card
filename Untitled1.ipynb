```python
import cv2
import os
from ultralytics import YOLO

# Load model YOLO detect ngÆ°á»i
model = YOLO("yolov8n.pt")

# ÄÆ°á»ng dáº«n video vÃ  thÆ° má»¥c lÆ°u áº£nh
video_path = "/content/drive/MyDrive/Myproject/ TEST!.mp4"
save_dir = "/content/drive/MyDrive/Myproject/box_nguoi_tung_frame"
os.makedirs(save_dir, exist_ok=True)

# Má»Ÿ video
cap = cv2.VideoCapture(video_path)
frame_idx = 0
saved_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame_idx += 1

    results = model(frame, classes=[0], conf=0.45)[0]  # Chá»‰ detect ngÆ°á»i (class 0)
    boxes = results.boxes

    if boxes is not None and len(boxes) > 0:
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            person_crop = frame[y1:y2, x1:x2]

            filename = os.path.join(save_dir, f"frame{frame_idx}_person{i+1}.jpg")
            cv2.imwrite(filename, person_crop)
            saved_count += 1

    print(f"ğŸŸ¢ Frame {frame_idx} | NgÆ°á»i phÃ¡t hiá»‡n: {len(boxes) if boxes else 0}", end="\r")

cap.release()
print(f"\nâœ… ÄÃ£ lÆ°u {saved_count} áº£nh bounding box ngÆ°á»i vÃ o: {save_dir}")

```

```python
import cv2
from ultralytics import YOLO
import os

# Load model cá»§a báº¡n
model_badge = YOLO('/content/drive/MyDrive/Myproject/weight_v1/train/weights/best.pt')

# Danh sÃ¡ch áº£nh cáº§n xá»­ lÃ½
image_paths = [
    "/content/frame110_person1.jpg",
    "/content/frame150_person2.jpg"
]

# TÃªn lá»›p: class 0 lÃ  dÃ¢y Ä‘eo (lanyard), class 1 lÃ  tháº» (card)
class_names = {0: "lanyard", 1: "card"}

# Táº¡o thÆ° má»¥c lÆ°u áº£nh káº¿t quáº£
output_dir = "/content/badge_detected_images"
os.makedirs(output_dir, exist_ok=True)

# Xá»­ lÃ½ tá»«ng áº£nh
for img_path in image_paths:
    img = cv2.imread(img_path)
    results = model_badge(img, conf=0.45)[0]

    if results.boxes is not None:
        for box in results.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls_id = int(box.cls[0])
            label = f"{class_names.get(cls_id, 'unknown')} {conf:.2f}"
            color = (0, 255, 0) if cls_id == 1 else (255, 255, 0)

            # Váº½ bounding box vÃ  nhÃ£n
            cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
            cv2.putText(img, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # LÆ°u áº£nh Ä‘Ã£ váº½ bbox
    filename = os.path.basename(img_path)
    cv2.imwrite(os.path.join(output_dir, filename), img)
    print(f"âœ… ÄÃ£ xá»­ lÃ½ vÃ  lÆ°u áº£nh: {filename}")

print("ğŸ¯ HoÃ n táº¥t. áº¢nh káº¿t quáº£ náº±m trong:", output_dir)
```

```python
import cv2
from ultralytics import YOLO
import numpy as np
import os

# Load models
model_person = YOLO("yolov8n.pt")  # NgÆ°á»i
model_badge = YOLO('/content/drive/MyDrive/Myproject/weight_v1/train/weights/best.pt')  # Tháº» vÃ  dÃ¢y Ä‘eo

# ÄÆ°á»ng dáº«n video input/output
video_in = "/content/drive/MyDrive/Myproject/ TEST!.mp4"
video_out = "/content/output_with_bbox1.mp4"

# Má»Ÿ video input
cap = cv2.VideoCapture(video_in)
fps = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Video writer (ghi video output)
writer = cv2.VideoWriter(video_out, cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))

frame_idx = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame_idx += 1

    # Detect ngÆ°á»i (class 0)
    results_person = model_person.track(frame, conf=0.45, classes=[0], persist=True)[0]
    # Detect tháº» vÃ  dÃ¢y Ä‘eo (sá»­ dá»¥ng cof = 0.5 cho tháº»/dÃ¢y Ä‘eo)
    results_badge = model_badge(frame, conf=0.5)[0]

    # Váº½ bounding box ngÆ°á»i
    if results_person.boxes is not None:
        for box in results_person.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cv2.rectangle(frame, (x1, y1), (x2, y2), (200, 200, 200), 2)

    # Váº½ bounding box tháº» vÃ  dÃ¢y Ä‘eo
    if results_badge.boxes is not None:
        for box in results_badge.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            label = "lanyard" if cls_id == 0 else "card"
            color = (255, 255, 0) if cls_id == 0 else (0, 255, 0)
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, f"{label} {conf:.2f}", (x1, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Ghi frame ra video output
    writer.write(frame)
    print(f"ğŸŸ¢ Frame {frame_idx} Ä‘Ã£ Ä‘Æ°á»£c xá»­ lÃ½", end="\r")

# Giáº£i phÃ³ng tÃ i nguyÃªn
cap.release()
writer.release()

print("\nâœ… Xá»­ lÃ½ hoÃ n táº¥t.")
print("ğŸ“‚ Video káº¿t quáº£ lÆ°u táº¡i:", video_out)
```
